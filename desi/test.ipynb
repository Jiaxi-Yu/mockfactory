```python
from matplotlib import pyplot as plt

for region in ['N', 'DN', 'DS']:
    catalog = Catalog.read('_tests/data-cutsky-y1-dark-3-{}.fits'.format(region))
    plt.scatter(catalog['RA'], catalog['DEC'])
plt.show()
```

```python
for region in ['N', 'DN', 'DS']:
    bins = np.linspace(0.8, 2.65, 100)
    catalog = Catalog.read('_tests/data-cutsky-y1-dark-3-{}.fits'.format(region))
    plt.hist(catalog['Z'], density=True, histtype='step', bins=bins)
    catalog = Catalog.read('_tests/randoms-cutsky-y1-dark-3-{}-0.fits'.format(region))
    plt.hist(catalog['Z'], density=True, histtype='step', bins=bins)
    plt.show()
```
